```python
# Standard imports
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
```

```python
# Read images
src = cv2.imread("/poissonblendingimages/source.jpg",cv2.IMREAD_UNCHANGED)
dst = cv2.imread("/poissonblendingimages/target.jpg", cv2.IMREAD_UNCHANGED)
cv2_imshow(src)
```

```python

# Create a rough mask around the airplane.
src_mask = np.ones(src.shape, dtype=np.uint8)*255
# src_mask[:,:] = True
cv2_imshow(src_mask)
```

```python
center = (dst.shape[1]//2, dst.shape[0]//2)
dst.shape
```

```python
output = cv2.seamlessClone(src, dst, src_mask, center, 0)#cv2.MIXED_CLONE)
```
